```python
def combine_years(data, cols):
    df = pd.DataFrame()
    for item, col in zip(data,cols):
        df[col] = item
    return df

def select_keys(df,index_col,column):
    data = []
    data.append(df[0][index_col].values)
    for i in range(len(df)):
        data.append(df[i][column].values.T)
    return data

def convert_to_int(df):
    for item in df:
        df[item] = df[item].astype(int)
    return df
```

```python
def rental_sqm(origin):
    cols = ['Region', '2012', '2013', '2014', '2015', '2016']
    rental_sqm = []
    for i in range(12,17):
        rental_sqm.append(pd.read_csv(data_folder+'rental_demographics20'+str(i)+'.csv').drop('Unnamed: 0', axis=1))
    data = []
    data.append(rental_sqm[0]['Unnamed: 0.1'].values)
    if origin == 'Swiss':
        for i in range(len(rental_sqm)):
            data.append(rental_sqm[i]['ohne Migrations-hintergrund'].values.T)
    if origin == 'Foreign':
        for i in range(len(rental_sqm)):
            data.append(rental_sqm[i]['mit Migrations-hintergrund 4)'].values.T)
    
    df = pd.DataFrame()
    for item, col in zip(data,cols):
        df[col] = item
        if 'X' in df[col].values:
            df[col][df[col] == 'X'] = None
            df[col] = df[col].astype(float)
    df = df.set_index('Region')
    return df
```

```python
swiss_sqm_df = rental_sqm('Swiss')
foreign_sqm_df = rental_sqm('Foreign')
```

```python
def get_region_sqm(df,region):
    rows = 7
    new_df = pd.DataFrame()
    for i, item in enumerate(df.index):
        if item == region:
            for j in range(1,rows):
                new_df = pd.concat([new_df,df.iloc[[i+j]] ])                       
    new_df.index.names = ['household_type']
    return new_df
```

```python
def household_reg(swiss_df, foreign_df, region_):
    hh_types = ['Einpersonenhaushalte','Paare ohne Kinder','Paare mit Kind','Einelternhaushalte',
               'Nichtfamilienhaushalte mit mehreren Personen','Mehrfamilienhaushalte']
    swiss = get_region_sqm(swiss_df,region_)
    foreign = get_region_sqm(foreign_df,region_)
    xi = np.arange(len(swiss.columns))
    lines_swiss = []
    lines_for = []
    for hh in hh_types:
        slope_, intercept_, r_value, p_value, std_err = stats.linregress(xi,swiss.loc[swiss.index.str.contains(hh)])
        lines_swiss.append(intercept_ + slope_*xi)
        slope, intercept, r_value, p_value, std_err = stats.linregress(xi,foreign.loc[foreign.index.str.contains(hh)])
        lines_for.append(intercept + slope*xi)
    
    colors = []
    for i in range(len(hh_types)):
        colors.append(np.random.rand(3))
    plt.figure()
    plt.subplot(2,2,1)
    for line_, hh_, color_ in zip(lines_swiss, hh_types, colors):
        plt.plot(line_, c=color_, label=hh_+' | Swiss')
        plt.legend()
        plt.title(region_)
    plt.subplot(2,2,2)
    for line, hh, color in zip(lines_for, hh_types, colors):
        plt.plot(line, c=color, label=hh+' | Foreign')
        plt.legend()
        plt.title(region_)
```

```python
region = ['Total','Genferseeregion','Espace Mittelland','Nordwestschweiz',
               'Zürich','Ostschweiz','Zentralschweiz','Tessin']
for reg in region:
    household_reg(swiss_sqm_df,foreign_sqm_df, str(reg))
```

```python
def drop_index(df, ind, rows = True):
    if len(df[0].columns) > len(ind):
        if rows:
            for i in range(len(df)):
                df[i] = df[i].drop(ind).reset_index().drop('index', 1)
        else:
            for i in range(len(df)):
                df[i] = df[i].drop(df[i].columns[ind], axis=1)
        return df
    return df

def select_rows(df,from_col,row_list):
    df = df[df[from_col].str.contains(row_list)].set_index(from_col)
    return df
```
